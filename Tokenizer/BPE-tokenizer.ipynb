BPE Tokenizer was used to train GPT-2 and GPT-3

Types of tokenization algorithms:
1. Word based 
2. Sub-word based
3. Character based 

1. Problem with word based tokenizer:
- Out of vocab (OOV) words. ( DIfficult to deal with)
- Different meaning of similar words. [boy, boys]

3. Character based tokenizer:
- Very small vocab, since every language has fixed number of letters/ characters. (English ~ 256)
- This solves the OOV problem  
Problem:
- Meaning associated with words is completely lost.
- Tokenized sequence iw longer than the raw text. { 'dinosaur' : ['d','i','n','o','s','a','u','r'] - 8 tokens}

Both 1 and 3 have the same issue of not capturing meaning of similiar words

2. Sub-word based tokenization:
RULES:
 1. Do not split frequently used words into subwords.
 2, Split rare words into smaller, meaningful subwords.
 eg: "boy" shouldn't be split, 
    "boys" should be split into "boy" and "s"

Advantages:
1. Helps model learn that different words with same root-word have same meaning.
2. It also helps identify suffixes and their usage contexts. Like in "Tokenization" and "Modernization" here the meaning is different but the suffix "ization" are used in similiar syntactic situations.


BYTE PAIR ENCODING is a subword tokenization algorithm.

BPE algorithm: Most common pair of consecutive bytes of data is replaced with byte that does not occur in data.

data: aaabdaaabac
Compressing - 
1. most common "byte-pair" will be replaced with Z as it doesnt exist in data 
data: ZabdZabac -> ZYdZYac
Compressed data: ZYdZYac
We stop here since only one pair remains and it only appars once, hence its not encoded - ac 

BPT ensures that most common words are represented as a single token, while rare words are broken down into two or more subword tokens.

Given a dataset - add /w at the end of every word
1. Split words into their charactes and get their frequencies ,/w is counted as seaparte character.
2. Find the byte-pair that occured most.
3. Merge the most frequent pairing.
4. Subtract the frequency of the byte-pair from the frequency of the individual elements.
5. This merged byte-paur is added as a new element in the dict.
6. In the next iteration find next ferquently occuring byte-pair.
7. Stopping criteria cound be no. of iterations or token count.
8. Finally remove all the element where frequency is 0. 

Key idea:
- Since /w here helps identify suffixes.
- eg: fin-est /w, easi-est/w, etc these words are always followed by /w, this helps the model learn that the token "est" mostly appears at the end as a suffix of root words.

BPE helps capture this relation


Advantage of Byte-pair:
1. It sovles OOV problem
2. It helps capture meaning of root words.

```python
import importlib
import tiktoken

importlib.metadata.version("tiktoken")
```

```python
# Vocabulary size of tokenization scheme used in gpt2 & gpt3 is - 50257
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = (
    "Hello, do you like tea? <|endoftext|> In the sunlit terraces"
    "of someunknownPlace."
)

integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})

print(integers)
```

```python
strings = tokenizer.decode(integers)

strings
```
