BPE Tokenizer was used to train GPT-2 and GPT-3

Types of tokenization algorithms:
1. Word based 
2. Sub-word based
3. Character based 

1. Problem with word based tokenizer:
- Out of vocab (OOV) words. ( DIfficult to deal with)
- Different meaning of similar words. [boy, boys]

3. Character based tokenizer:
- Very small vocab, since every language has fixed number of letters/ characters. (English ~ 256)
- This solves the OOV problem  
Problem:
- Meaning associated with words is completely lost.
- Tokenized sequence iw longer than the raw text. { 'dinosaur' : ['d','i','n','o','s','a','u','r'] - 8 tokens}

Both 1 and 3 have the same issue of not capturing meaning of similiar words

2. Sub-word based tokenization:
RULES:
 1. Do not split frequently used words into subwords.
 2, Split rare words into smaller, meaningful subwords.
 eg: "boy" shouldn't be split, 
    "boys" should be split into "boy" and "s"

Advantages:
1. Helps model learn that different words with same root-word have same meaning.
2. It also helps identify suffixes and their usage contexts. Like in "Tokenization" and "Modernization" here the meaning is different but the suffix "ization" are used in similiar syntactic situations.


BYTE PAIR ENCODING is a subword tokenization algorithm.
